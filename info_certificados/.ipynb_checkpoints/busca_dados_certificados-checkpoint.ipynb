```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import datetime
import time
import numpy as np
from fake_useragent import UserAgent

def verifica_request(req):
    
    print(req.url)
    # Verificando se a request deu certo
    if req.status_code != requests.codes.ok:
        erro = f'''
        Error code: {req.status_code}
        Reason: {req.reason}
        Message: {req.content}
        '''
        raise ValueError(erro)
        
    else:
        print('OK...')

def busca_html(url):

    headers = {
        'User-Agent': UserAgent().random
    }

    req = requests.request(method='GET', url=url, headers=headers)
    verifica_request(req)

    soup = BeautifulSoup(req.text, 'html.parser')
    return soup
```

## Buscando as informações dos certificados

```python
url = 'https://cursos.alura.com.br/user/viniantunes'
soup = busca_html(url)

cursos_concluidos = soup.findAll('li', {'class': 'card-list__item'})
```

## Organizando um DataFrame com as informações buscadas

```python
saida = pd.DataFrame()
cursos = {}

for curso in cursos_concluidos:

    cursos['titulo'] = curso.attrs['data-course-name']
    cursos['data_inicio'] = datetime.datetime.strptime(curso.attrs['data-started-at'], '%m/%d/%Y').date()
    cursos['data_fim'] = datetime.datetime.strptime(curso.attrs['data-finished-at'], '%m/%d/%Y').date()
    cursos['certificado'] = url.split('/user/viniantunes')[0] + curso.find('a', {'class': 'course-card__certificate bootcamp-text-color'}).attrs['href']
    cursos['icone'] = curso.find('img').attrs['src']
    df_cursos = pd.DataFrame([cursos])
    saida = pd.concat([saida, df_cursos], ignore_index=True)
saida.sort_values(by='data_fim', inplace=True, ignore_index=True)
```

## O DataFrame ficou assim...

```python
dados = saida.copy()
dados
```

# Buscando mais informações de cada curso

```python
%%time
carga_horaria = []
exercicios = []
certficate_auths = []
cod_auths = []

for i, url_certificado in enumerate(dados['certificado']):
    
    tempo_espera = np.random.randint(2, 5)
    time.sleep(tempo_espera)
    print(f'{i+1}/{len(dados)}', end=' --> ')
    infos = busca_html(url_certificado)
    
    carga_horaria.append(int(infos.find('span', {'class': 'certificate-hours'}).getText().split(' horas')[0].split('estimada em ')[1]))
    exercicios.append(int(infos.find('span', {'class': 'exercises-done'}).getText().split(' de ')[0].strip()))
    certficate_auths.append(infos.find('a', {'class': 'authenticity alura'}).getText())
    cod_auths.append(infos.find('a', {'class': 'authenticity alura'}).getText().split('/')[-1])

dados['carga_horaria'] = carga_horaria
dados['exercicios'] = exercicios
dados['certficate_auths'] = certficate_auths
dados['cod_auths'] = cod_auths
```

## Preparando o html para a página de certificados
##### Fiz isso pois o GitHub pages não aceita o JavaScript

```python
def adiciona_icone(titulo_curso, link_certificado, link_icone):
    
    novo_icone = ""
    novo_icone += "\t\t\t<span data-tooltip='" + titulo_curso + "'>\n"
    novo_icone += "\t\t\t\t<a href='" + link_certificado + "'>\n"
    novo_icone += "\t\t\t\t<img src='" + link_icone + "' alt='" + titulo_curso + "'>\n"
    novo_icone += "\t\t\t\t</a>\n"
    novo_icone += "\t\t\t</span>\n\n"
    return novo_icone
```

```python
html = ""
for titulo_curso, link_certificado, link_icone in zip(dados['titulo'], dados['certficate_auths'], dados['icone']):
    html += adiciona_icone(titulo_curso, link_certificado, link_icone)
```

## Salvando o html num arquivo texto

```python
with open('data\\trecho_html.txt', 'w') as arquivo:
    arquivo.write(html)
```
